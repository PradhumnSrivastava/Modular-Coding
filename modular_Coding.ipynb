## **Modular Coding**

### What is Modular Coding?

Modular coding is a programming approach where a large project is divided into **small independent modules**.

Each module performs a specific task such as:
- Data Processing
- Model Training
- Evaluation
- Utilities

This makes the project more organized and manageable.

---

### Example

#### Without Modular Coding

```python
load_data()
clean_data()
train_model()
evaluate_model()
```

#### With Modular Coding

```python
### data_loader.py
load_data()
```

```python
### trainer.py
train_model()
```

```python
### evaluation.py
evaluate_model()
```

---

### Objective of Modular Coding

- Improve code readability
- Make code reusable
- Build scalable systems
- Simplify debugging
- Support team collaboration

---

### Importance in AI & ML

Machine Learning projects contain many components like:
- Data Pipelines
- Feature Engineering
- Model Training
- Deployment

Modular coding helps organize these components efficiently.

#### **Data Ingestion**

Data Ingestion is the process of **collecting and importing data** from different sources into a system for further processing and analysis.

It is the first step in most:
- Machine Learning
- Deep Learning
- AI pipelines


#### **Purpose of Data Ingestion**

- Collect raw data
- Load data into the project
- Prepare data for preprocessing
- Automate data flow
