## Universal File Structure for Data Science & AI Projects

```text
project_name/
│
├── data/
│   ├── raw/
│   ├── processed/
│   └── external/
│
├── notebooks/
│
├── src/
│   ├── data_ingestion/
│   ├── preprocessing/
│   ├── feature_engineering/
│   ├── model_training/
│   ├── model_evaluation/
│   ├── prediction/
│   ├── visualization/
│   ├── pipelines/
│   └── utils/
│
├── configs/
│
├── models/
│
├── logs/
│
├── artifacts/
│
├── tests/
│
├── requirements.txt
├── setup.py
├── .gitignore
├── README.md
└── main.py
```

---

## Folder Explanation

### `data/`
Stores datasets.

- `raw/` → Original data
- `processed/` → Cleaned data
- `external/` → Third-party data

---

### `notebooks/`
Jupyter notebooks for:
- EDA
- experimentation
- research

---

### `src/`
Main source code of the project.

Contains:
- data ingestion
- preprocessing
- training
- evaluation
- pipelines
- utilities

---

### `configs/`
Configuration files:
- paths
- hyperparameters
- model settings

---

### `models/`
Saved trained models.

Example:
- `.pkl`
- `.pt`
- `.h5`

---

### `logs/`
Stores logs and debugging information.

---

### `artifacts/`
Generated outputs:
- processed datasets
- reports
- metrics
- plots

---

### `tests/`
Testing files for validating modules.

---

### `requirements.txt`
Project dependencies.

Example:

```text
pandas
numpy
scikit-learn
matplotlib
```

---

### `README.md`
Project documentation.

---

### `main.py`
Main execution file of the project.

---

## Final Thought

> “A good project structure makes AI systems scalable, maintainable, and production-ready.”
